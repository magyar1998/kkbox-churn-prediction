```python
# Configurações
arquivo_origem = "/content/drive/Shareddrives/Real_Project_Churn_DSA30/Dataset/Raw/kkboxs/user_logs.csv"
folder_output = "/content/drive/Shareddrives/Real_Project_Churn_DSA30/Dataset/Raw/kkboxs/user_logs"
tamanho_chunk = 10_000_000
contador = 1

# Loop para ler e salvar em partes menores
for chunk in pd.read_csv(arquivo_origem, chunksize=tamanho_chunk):
    nome_saida = f"parte_{contador}.csv"
    chunk.to_csv(os.path.join(folder_output, nome_saida), index=False)
    print(f"Arquivo {nome_saida} salvo com {len(chunk)} linhas.")
    contador += 1
```
